```python
!pip install google-colab
import numpy as np
from ultralytics import YOLO
import cv2
import math

def pose_estimation(filename_in, filename_out, model):
    """ 画像ファイルから人間を検出して姿勢推定する関数 """

    # 画像をオープン
    img = cv2.imread(filename_in, 1)

    # 画像を表示
    #cv2.imshow('img', img)
    cv2.waitKey()
    cv2.destroyAllWindows()

    # 物体検出
    results = model(img,save=True, save_crop=True)

    # バウンディングボックスをオーバーレイTrue
    img_annotated = results[0].plot(boxes=False)
    results2 = model(img_annotated,save=True, save_crop=True)
    # 物体検出結果画像を表示
    #cv2.imshow("pose-estimated", img_annotated)
    cv2.waitKey()

    # 画像を保存[0][5,1].item()
    cv2.imwrite(filename_out, img_annotated)
    print(results[0].keypoints.xy)
    xys = results[0].keypoints.xy
    print(xys[0][5,0])
    for i in range(results[0].keypoints.xy.size()[0]):
        print(i)
        a=np.sqrt((xys[i][5,0].item()-xys[i][6,0].item())**2+(xys[i][5,1].item()-xys[i][6,1].item())**2)
        c=np.sqrt((xys[i][5,0].item()-xys[i][3,0].item())**2+(xys[i][5,1].item()-xys[i][3,1].item())**2)
        b=np.sqrt((xys[i][6,0].item()-xys[i][3,0].item())**2+(xys[i][3,1].item()-xys[i][6,1].item())**2)
        ls=(a+b+c)/2
        s=np.sqrt(ls*(ls-a)*(ls-b)*(ls-c))

        if (a>=b and a>=c):
            h=2*s/a
            lb=math.degrees(math.asin(h/c))
        elif (b>=a and b>=c):
            h=2*s/b
            lb=180-math.degrees(math.asin(h/c))-math.degrees(math.asin(h/a))
        elif (c>=a and c>=b):
            h=2*s/c
            lb=math.degrees(math.asin(h/a))
        if(xys[i][5,0].item()!=0 and xys[i][5,1].item()!=0 and xys[i][6,0].item()!=0 and xys[i][6,1].item()!=0 and xys[i][3,0].item()!=0 and xys[i][3,1].item()!=0):
            print(lb)
        a=np.sqrt((xys[i][5,0].item()-xys[i][6,0].item())**2+(xys[i][5,1].item()-xys[i][6,1].item())**2)
        c=np.sqrt((xys[i][5,0].item()-xys[i][4,0].item())**2+(xys[i][5,1].item()-xys[i][4,1].item())**2)
        b=np.sqrt((xys[i][6,0].item()-xys[i][4,0].item())**2+(xys[i][4,1].item()-xys[i][6,1].item())**2)
        ls=(a+b+c)/2
        s=np.sqrt(ls*(ls-a)*(ls-b)*(ls-c))

        if (a>=b and a>=c):
            h=2*s/a
            lc=math.degrees(math.asin(h/b))
        elif (b>=a and b>=c):
            h=2*s/b
            lc=math.degrees(math.asin(h/a))
        elif (c>=a and c>=b):
            h=2*s/c
            lc=180-math.degrees(math.asin(h/b))-math.degrees(math.asin(h/a))
        if(xys[i][5,0].item()!=0 and xys[i][5,1].item()!=0 and xys[i][6,0].item()!=0 and xys[i][6,1].item()!=0 and xys[i][4,0].item()!=0 and xys[i][4,1].item()!=0):
            print(lc)



if __name__ == '__main__':
    """ Main """

    filename_in = 'a6-6-8.jpg'
    filename_out = 'pose-out.jpg'

    # モデルを設定
    model = YOLO('yolov8s-pose.pt')

    # 物体検出関数を実行
    pose_estimation(filename_in, filename_out, model)
```
